```python
import pandas as pd
import os
```

```python
parquet_list = []
for file in os.listdir():
    if file.endswith(".parquet"):
        parquet_list.append(file)
        print(parquet_list)
```

```python
dfs = {}
for parquet in parquet_list:
    dfs[f"{parquet.removesuffix(".parquet")}_df"] = pd.read_parquet(parquet)
    print("Done!")
```

## Question 3. Counting short trips

For the trips in November 2025 (lpep_pickup_datetime between '2025-11-01' and '2025-12-01', exclusive of the upper bound), how many trips had a `trip_distance` of less than or equal to 1 mile?

```python
november_trips = dfs['green_tripdata_2025-11_df'][dfs['green_tripdata_2025-11_df']['lpep_pickup_datetime'].between('2025-11-01','2025-12-01',inclusive='left')]
november_trips
```

```python
trips_with_less_or_equal_1_mile = november_trips[november_trips['trip_distance']<=1]
trips_with_less_or_equal_1_mile
```

```python
print(f"Count of November trips with less than or equal to 1 mile : {trips_with_less_or_equal_1_mile.shape[0]}")
```

## Question 4. Longest trip for each day

Which was the pick up day with the longest trip distance? Only consider trips with `trip_distance` less than 100 miles (to exclude data errors).

Use the pick up time for your calculations.

```python
trips_less_than_100_miles = dfs['green_tripdata_2025-11_df'][dfs['green_tripdata_2025-11_df']['trip_distance'] < 100]
trips_less_than_100_miles
```

```python
day_with_longest_trip_distance = trips_less_than_100_miles.sort_values(by='trip_distance',ascending=False).iloc[0:1]
day_with_longest_trip_distance
```

```python
print(f"The pick up day with longest trip distance was : {day_with_longest_trip_distance['lpep_pickup_datetime'].dt.date.squeeze()}")
```

## Question 5. Biggest pickup zone

Which was the pickup zone with the largest `total_amount` (sum of all trips) on November 18th, 2025?

```python
november_18th_trips = dfs['green_tripdata_2025-11_df'][dfs['green_tripdata_2025-11_df']['lpep_pickup_datetime'].dt.date.astype(str) == '2025-11-18']
november_18th_trips
```

```python
pickup_zones_with_total_amount = november_18th_trips.groupby(by='PULocationID',as_index=False)['total_amount'].sum()
pickup_zones_with_total_amount
```

```python
pickup_zone_with_highest_amount = pickup_zones_with_total_amount.sort_values(by='total_amount',ascending=False).iloc[0:1]
pickup_zone_with_highest_amount
```

```python
zone_df = pd.read_csv('taxi_zone_lookup.csv')
zone_df
```

```python
location_id_of_pickup_zone = pickup_zone_with_highest_amount['PULocationID'].squeeze()
location_id_of_pickup_zone
```

```python
pickup_zone_with_highest_amount = zone_df[zone_df['LocationID'] == location_id_of_pickup_zone]
pickup_zone_with_highest_amount
```

```python
print(f"The pickup zone with the largest total amount (sum of all trips) on November 18th, 2025 is : {pickup_zone_with_highest_amount['Zone'].squeeze()}")
```

## Question 6. Largest tip

For the passengers picked up in the zone named "East Harlem North" in November 2025, which was the drop off zone that had the largest tip?

Note: it's `tip` , not `trip`. We need the name of the zone, not the ID.

```python
east_harlem_north_location_id = zone_df[zone_df['Zone'] == 'East Harlem North']['LocationID'].squeeze()
east_harlem_north_location_id
```

```python
passengers_picked_up_in_east_harlem_north = dfs['green_tripdata_2025-11_df'][dfs['green_tripdata_2025-11_df']['PULocationID'] == east_harlem_north_location_id]
passengers_picked_up_in_east_harlem_north
```

```python
passengers_picked_up_in_east_harlem_north_november = passengers_picked_up_in_east_harlem_north[passengers_picked_up_in_east_harlem_north['lpep_pickup_datetime'].between('2025-11-01','2025-12-01',inclusive='left')]
passengers_picked_up_in_east_harlem_north_november
```

```python
dropoff_zone_with_largest_tip_amount = passengers_picked_up_in_east_harlem_north_november.sort_values(by="tip_amount",ascending=False).iloc[0:1]
dropoff_zone_with_largest_tip_amount
```

```python
dropoff_zone_with_largest_tip_amount_full_details = zone_df[zone_df['LocationID'] == dropoff_zone_with_largest_tip_amount['DOLocationID'].squeeze()]
dropoff_zone_with_largest_tip_amount_full_details
```

```python
print(f"For the passengers picked up in the zone named `East Harlem North` in November 2025, the drop off zone that had the largest tip was : {dropoff_zone_with_largest_tip_amount_full_details['Zone'].squeeze()}")
```
